## **3. ÄÃ¡nh giÃ¡ mÃ´ hÃ¬nh**

Sau khi huáº¥n luyá»‡n mÃ´ hÃ¬nh xong, cÃ¡c thÃ­ sinh cáº§n:

1. **Dá»± Ä‘oÃ¡n trÃªn táº­p validation** vÃ  lÆ°u káº¿t quáº£
2. **TÃ­nh toÃ¡n cÃ¡c metrics** Ä‘Ã¡nh giÃ¡ chÃ­nh:
   - **Dice Coefficient**: Äo Ä‘á»™ tÆ°Æ¡ng Ä‘á»“ng giá»¯a prediction vÃ  ground truth
   - **IoU (Intersection over Union)**: Tá»· lá»‡ giao vÃ  há»£p cá»§a hai mask
   - **F1-Score**: Harmonic mean cá»§a precision vÃ  recall

3. **Táº¡o visualization** cho 5 samples ngáº«u nhiÃªn tá»« táº­p validation
4. **Implement XAI** Ä‘á»ƒ táº¡o attention heatmap

### CÃ´ng thá»©c tÃ­nh metrics:

**Dice Coefficient:**
$$Dice = \frac{2|A \cap B|}{|A| + |B|}$$

**IoU:**
$$IoU = \frac{|A \cap B|}{|A \cup B|}$$

**F1-Score:**
$$F1 = \frac{2 \times Precision \times Recall}{Precision + Recall}$$

### YÃªu cáº§u visualization:
- Hiá»ƒn thá»‹ áº£nh gá»‘c, ground truth mask, predicted mask, vÃ  overlay
- Táº¡o heatmap XAI (GradCAM) Ä‘á»ƒ minh hoáº¡ vÃ¹ng chÃº Ã½ cá»§a model
- LÆ°u 5 káº¿t quáº£ visualization tá»‘t nháº¥t vÃ o `/kaggle/working/predictions/`
- So sÃ¡nh prediction vá»›i ground truth trÃªn validation set


````python
# VÃ­ dá»¥ code Ä‘Ã¡nh giÃ¡ metrics vÃ  visualization
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os

def calculate_dice(pred_mask, true_mask):
    """TÃ­nh Dice Coefficient"""
    intersection = np.logical_and(pred_mask, true_mask).sum()
    union = pred_mask.sum() + true_mask.sum()
    dice = (2.0 * intersection) / union
    return dice

def calculate_iou(pred_mask, true_mask):
    """TÃ­nh IoU"""
    intersection = np.logical_and(pred_mask, true_mask).sum()
    union = np.logical_or(pred_mask, true_mask).sum()
    iou = intersection / union
    return iou

def calculate_f1(pred_mask, true_mask):
    """TÃ­nh F1-Score"""
    precision = np.logical_and(pred_mask, true_mask).sum() / pred_mask.sum()
    recall = np.logical_and(pred_mask, true_mask).sum() / true_mask.sum()
    f1 = 2 * precision * recall / (precision + recall)
    return f1

# VÃ­ dá»¥ visualization vá»›i dataset Ä‘Ã£ xá»­ lÃ½
def visualize_results(image, true_mask, pred_mask, save_path=None):
    """Táº¡o visualization káº¿t quáº£"""
    fig, axes = plt.subplots(1, 4, figsize=(16, 4))
    
    # áº¢nh gá»‘c
    axes[0].imshow(image)
    axes[0].set_title('áº¢nh gá»‘c')
    axes[0].axis('off')
    
    # Ground truth mask
    axes[1].imshow(true_mask, cmap='gray')
    axes[1].set_title('Ground Truth')
    axes[1].axis('off')
    
    # Predicted mask
    axes[2].imshow(pred_mask, cmap='gray')
    axes[2].set_title('Prediction')
    axes[2].axis('off')
    
    # Overlay prediction trÃªn áº£nh gá»‘c
    overlay = image.copy()
    overlay[pred_mask > 0.5] = [255, 0, 0]  # Äá» cho prediction
    axes[3].imshow(overlay)
    axes[3].set_title('Overlay')
    axes[3].axis('off')
    
    # TÃ­nh metrics
    dice = calculate_dice(pred_mask > 0.5, true_mask > 0.5)
    iou = calculate_iou(pred_mask > 0.5, true_mask > 0.5)
    f1 = calculate_f1(pred_mask > 0.5, true_mask > 0.5)
    
    # ThÃªm metrics vÃ o title
    fig.suptitle(f'Dice: {dice:.3f}, IoU: {iou:.3f}, F1: {f1:.3f}', fontsize=14)
    
    plt.tight_layout()
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"Visualization Ä‘Ã£ Ä‘Æ°á»£c lÆ°u táº¡i: {save_path}")
    
    plt.show()

# VÃ­ dá»¥ sá»­ dá»¥ng vá»›i dataset Ä‘Ã£ xá»­ lÃ½
print("ðŸ“Š VÃ­ dá»¥ sá»­ dá»¥ng metrics vÃ  visualization:")
print("âœ… CÃ¡c hÃ m calculate_dice, calculate_iou, calculate_f1 Ä‘Ã£ Ä‘Æ°á»£c Ä‘á»‹nh nghÄ©a")
print("âœ… HÃ m visualize_results Ä‘Ã£ Ä‘Æ°á»£c Ä‘á»‹nh nghÄ©a")
print("ðŸŽ¯ Sáºµn sÃ ng Ä‘á»ƒ sá»­ dá»¥ng vá»›i dataset Ä‘Ã£ xá»­ lÃ½ trÃªn Kaggle!")
````
